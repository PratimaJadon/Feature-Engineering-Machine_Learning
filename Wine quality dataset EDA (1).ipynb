Fixed acidity: The amount of nonvolatile acids in the wine.

Volatile acidity: The amount of acetic acid in the wine, which can contribute to vinegar-like flavors.

Citric acid: The amount of citric acid in the wine, which can add freshness and citrus flavors.

Residual sugar: The amount of sugar left in the wine after fermentation is complete.

Chlorides: The amount of salt in the wine, which can impact its taste.

Free sulfur dioxide: The amount of free form sulfur dioxide, an antimicrobial and antioxidant, in the wine.

Total sulfur dioxide: The total amount of sulfur dioxide present in the wine.

Density: The density of the wine, which is related to its alcohol content and sugar concentration.

pH: The acidity or basicity of the wine on a logarithmic scale.

Sulphates: The amount of sulfur dioxide bound with other chemicals in the wine.

Alcohol: The alcohol percentage of the wine.

Quality: A rating or score given to the wine by experts or consumers, typically on a scale from 0 to 10.
